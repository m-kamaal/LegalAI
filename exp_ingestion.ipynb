```python
from pypdf import PdfReader

def load_doc(doc_path):
    text=""
    if ".pdf" in doc_path:
        reader = PdfReader(doc_path)
        for page in reader.pages:
            text+= page.extract_text()
    elif ".txt" in doc_path:
        with open(doc_path, "r") as f:
            text = f.read()
    return text

load_doc("/Users/workpc/Legalai/harry_potter.txt")
```

### Extract various metadata and text from docling file

```python
def extract_text_blocks_from_texts(doc2):    
    result = {}

    result["document_info"] = {
        "doc_name": doc2.name,
        "doc_type": getattr(doc2.origin, 'mimetype'),
        "doc_version": doc2.version
    }


    result["blocks"] = []
    for item in doc2.texts:
        text = getattr(item, "text", "").strip()
        if text:
            meta = {
                "page_number": item.prov[0].page_no if item.prov else None,
                "content_label": getattr(item, "label", None).name if getattr(item, "label", None) else None,
                # "bbox": item.prov[0].bbox if item.prov else None,
                # "char_span": item.prov[0].charspan if item.prov else None,
                "orig_text": getattr(item, "orig", ""),
                "level": getattr(item, "level", None),  # section header level if present
            }
            result["blocks"].append({"text": text, "metadata": meta})
    return result
```

```python

text_blocks = extract_text_blocks_from_texts(doc2)["blocks"]
document_information = extract_text_blocks_from_texts(doc2)["document_info"]
print(f"Name of the file: {document_information['doc_name']}")
print(f"Type of the file: {document_information['doc_type']} \n")
for b in text_blocks[:25]:
    print(f"Page: {b['metadata']['page_number']}, Label: {b['metadata']['content_label']}, Level: {b['metadata']['level']}")
    print(b["text"][:300])
    print("---")

```
